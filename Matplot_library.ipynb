```python
from matplotlib import pyplot as plt
```

```python
import seaborn as sns
```

```python
import pandas as pd
```

```python
data = sns.load_dataset("dots").head(50)
data
```

```python
x=data.iloc[:,3].values
x
```

```python
y = data.iloc[:,4].values
y
```

# Line PLot in Matplot

```python
plt.plot(x,y,marker='o',linestyle="dashed",ms=10,mec='r',mfc='g')
plt.title("Line_Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.grid()
plt.show()
```

# Bar plot in matplot

```python
plt.bar(x,y)
plt.grid()
plt.show()
```

# Histogram in matplot

```python
plt.hist(x)
plt.grid()
plt.plot()
plt.show()
```

# Scatter Plot in matplot

```python
plt.scatter(x,y,marker='*')

plt.grid()

plt.show()
```

```python
plt.plot(x,marker='o',linestyle="dashed",label='x')
plt.plot(y,marker='*',label='y')
plt.grid()
plt.legend()
plt.show()
```

# Histogram plot in matplot

```python
plt.pie(y)
plt.pie(x)

plt.show()

```
