```python
from matplotlib import pyplot as plt
```

# Line plot in matplot....

```python
plt.plot([2,9,3],[9,4,6])
plt.title('Info')
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.show()
```

# Bar plot in matplot.....

```python
x = [2,9,4,3,4,8,3,1,5,3,9,6]
y = [2,9,4,3,8,7,2,4,6,5,3,3]
```

```python
plt.bar(x,y)
plt.title('Bar plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.plot()
plt.show()
```

# Histogram in matplot

```python
y = [9,7,3,1,6,4,5,3]
plt.hist(y)
plt.show()
```

# Scatter plot in matplot

```python
x = [8,6,4,9,2,4,7]
```

```python
y = [3,4,9,6,1,2,7]
```

```python
plt.scatter(x,y)
plt.plot()
plt.show()
```

# Plotting with marker and many other .....

```python
import numpy as np
x = np.array([3,4,9,7,2,8])
y = np.array([8,3,1,7,6,4])
plt.plot(x,y,marker='o')
plt.show()
```

```python
plt.plot(x,y,marker='>')
plt.show()
```

```python
plt.plot(x,y,marker='*')
plt.show()
```

```python
plt.plot(x,y,'o:r')
plt.show()
```

```python
plt.plot(x,y,marker='o',ms=12,mec='b',mfc='g',linestyle='dashed')
plt.show()
```

```python
plt.plot(x,y,marker='o',ms=12,mec='b',mfc='g',linestyle='dashed',color='r')
plt.show()
```

```python
plt.plot(x,marker='>')
plt.plot(y,linestyle='dashed',marker='o')
plt.grid()
plt.show()
```

# Pie Plot in matplot

```python
y = np.array([66,54,22,34,9,12])
plt.pie(y)
plt.plot()
plt.grid()
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
a = np.array([3,9,4,2,8,1,6])
b = np.array([9,2,7,3,1,6,4])

plt.plot(a,marker='o',linestyle=':',ms='15',mec='g',mfc='r',label='a')
plt.plot(b,marker='>',label='b',ms='15')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid()
plt.legend()
plt.show()
```
